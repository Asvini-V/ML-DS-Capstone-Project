### 1. Import Libraries 

```python
import pandas as pd
import numpy as np
import pickle
```

### 2. Load Model & Pre-Processing Objects

```python
model = pickle.load(open("Final_Model.pkl", "rb"))
selector = pickle.load(open("F_selector.pkl", "rb"))
scaler = pickle.load(open("scaler.pkl", "rb"))
model_columns = pickle.load(open("columns.pkl", "rb"))
```

### 3. Load Dataset

```python
dataset = pd.read_csv("preprocessed_data.csv")
X = dataset.drop('disease_diagnosis', axis=1)
print("\nEnter patient details:\n")
```

### 4. Take User Input

```python
# Numeric + binary inputs
data = {
    'bmi': float(input("BMI: ")),
    'alcohol_consumption': int(input("Alcohol (0=Heavy/1=Moderate/2=None): ")),
    'sun_exposure': int(input("Sun Exposure (0=Low/1=Moderate/2=High): ")),
    'income_level': int(input("Income Level (0=High/1=Low/2=Middle): ")),
    'latitude_region': int(input("Region (0=High/1=Low/2=Mid): ")),

    'vitamin_a_percent_rda': float(input("Vit A: ")),
    'vitamin_c_percent_rda': float(input("Vit C: ")),
    'vitamin_d_percent_rda': float(input("Vit D: ")),
    'vitamin_e_percent_rda': float(input("Vit E: ")),
    'vitamin_b12_percent_rda': float(input("Vit B12: ")),

    'folate_percent_rda': float(input("Folate: ")),
    'calcium_percent_rda': float(input("Calcium: ")),
    'iron_percent_rda': float(input("Iron: ")),

    'hemoglobin_g_dl': float(input("Hemoglobin: ")),
    'serum_vitamin_d_ng_ml': float(input("Serum Vit D: ")),
    'serum_vitamin_b12_pg_ml': float(input("Serum B12: ")),
    'serum_folate_ng_ml': float(input("Serum Folate: ")),

    'symptoms_count': int(input("Symptoms Count: ")),

    'has_night_blindness': int(input("Night Blindness (0=No/1=Yes): ")),
    'has_fatigue': int(input("Fatigue (0=No/1=Yes): ")),
    'has_bleeding_gums': int(input("Bleeding Gums (0=No/1=Yes): ")),
    'has_bone_pain': int(input("Bone Pain (0=No/1=Yes): ")),
    'has_muscle_weakness': int(input("Muscle Weakness (0=No/1=Yes): ")),
    'has_numbness_tingling': int(input("Numbness (0=No/1=Yes): ")),
    'has_memory_problems': int(input("Memory Issues (0=No/1=Yes): ")),
    'has_pale_skin': int(input("Pale Skin (0=No/1=Yes): "))
}

# Categorical inputs
data.update({
    "gender": input("Gender (Male/Female): "),
    "smoking_status": input("Smoking (Former/Never/Current): "),
    "exercise_level": input("Exercise (Light/Moderate/Sedentary): "),
    "diet_type": input("Diet (Pescatarian/Vegan/Vegetarian/Omnivore): ")
})
```

### 5. Input PreProcessing

```python
# Convert to DataFrame

input_df = pd.DataFrame([data])
```

```python
# One-Hot Encoding
input_encoded = pd.get_dummies(input_df)

# Match training columns
input_encoded = input_encoded.reindex(columns=model_columns, fill_value=0)
input_encoded = input_encoded.astype(int)
```

### 5. Feature Selection + Scaling

```python
# Feature Selection + Scaling
input_final = scaler.transform(selector.transform(input_encoded))
```

### 6. Prediction

```python
prediction = model.predict(input_final)[0]
```

```python
label_map = {
    0: "Anemia",
    1: "Healthy",
    2: "Night Blindness",
    3: "Rickets_Osteomalacia",
    4: "Scurvy"
}

print("\nPredicted Disease:", label_map.get(prediction, "Unknown"))
```
