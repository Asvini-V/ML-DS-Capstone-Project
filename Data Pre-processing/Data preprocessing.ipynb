```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

## 2. Load dataset

```python
raw_df=pd.read_csv("vitamin_deficiency_disease_dataset.csv")  # Read the raw dataset
raw_df
```

## 4. Data Pre-Processing

### 4.1. Drop Irrelevant Columns

```python
raw_df = raw_df.drop('has_multiple_deficiencies', axis=1)
raw_df = raw_df.drop('symptoms_list', axis=1)
```

### 4.2. Explore Categorical Features

```python
print(raw_df['gender'].value_counts())
print(raw_df['smoking_status'].value_counts())
print(raw_df['exercise_level'].value_counts())
print(raw_df['diet_type'].value_counts())
```

### 4.3. One-Hot Encoding (Categorical variables)

```python
# Convert categorical columns into numerical format
raw_df = pd.get_dummies(raw_df,columns=['gender', 'smoking_status', 'exercise_level', 'diet_type'],drop_first=True)   # avoid redundancy
raw_df
```

### 4.4. Label Encoding (Ordinal columns)

```python
le = LabelEncoder()
cols = ['alcohol_consumption', 'sun_exposure', 'income_level', 'latitude_region']
for col in cols:
    raw_df[col] = le.fit_transform(raw_df[col])
```

### 4.5. Encode Target Variable

```python
raw_df['disease_diagnosis'] = le.fit_transform(raw_df['disease_diagnosis'])
```

### 4.6. Convert Boolean to Integer

```python
# Convert True/False → 1/0
bool_cols = raw_df.select_dtypes(include='bool').columns
raw_df[bool_cols] = raw_df[bool_cols].astype(int)
```

### 4.7. Final Check

```python
print(raw_df.head())
print(raw_df.dtypes)
```
